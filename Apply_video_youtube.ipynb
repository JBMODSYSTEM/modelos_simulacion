```python
import pandas as pd  # Importamos la biblioteca pandas para trabajar con datos estructurados como DataFrames.
import numpy as np  # Importamos la biblioteca numpy para realizar operaciones matemáticas y de álgebra lineal.
import os as os  # Importamos la biblioteca os para interactuar con el sistema operativo.

# Carga un archivo CSV llamado 'data/pima-indians-diabetes.csv' en un DataFrame de pandas.
filename = 'data/pima-indians-diabetes.csv'
names = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome']
df = pd.read_csv(filename, names=names)  # Lee el archivo CSV y asigna nombres a las columnas.

# Muestra las primeras 5 filas del DataFrame para inspeccionar los datos.
df.head()
```
